```python
import pandas as pd

# Load the dataset with error handling
try:
    df = pd.read_csv("iris.csv")  # Replace with your dataset path
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("The file was not found. Please check the file path.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
print(df.head())  # First 5 rows
print(df.info())  # Structure and types
print(df.isnull().sum())  # Check for missing values
```

```python
# Check if there are any missing values and fill them if necessary
df.fillna(method='ffill', inplace=True)  # Forward fill if needed
```

```python
print(df.describe())
```

```python
grouped = df.groupby("variety").mean()
print(grouped)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
df.groupby("variety")["petal.length"].mean().plot(kind="bar", color="orange")
plt.title("Bar Chart: Average Petal Length by Variety")
plt.xlabel("Variety")
plt.ylabel("Average Petal Length")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df["petal.length"], bins=20, color="purple", alpha=0.7)
plt.title("Histogram: Distribution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["sepal.length"], df["petal.length"], c=df["variety"].map({"Setosa": "red", "Versicolor": "blue", "Virginica": "green"}), alpha=0.7)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style
sns.set(style="whitegrid")

# Create a bar plot of the average petal length by variety
plt.figure(figsize=(8, 6))
sns.barplot(x="variety", y="petal.length", data=df, palette="muted")
plt.title("Bar Chart: Average Petal Length by Variety")
plt.xlabel("Variety")
plt.ylabel("Average Petal Length")
plt.show()
```

```python
# Create a histogram of petal length
plt.figure(figsize=(8, 6))
sns.histplot(df["petal.length"], bins=20, kde=True, color="purple")
plt.title("Histogram: Distribution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot for sepal length vs petal length, colored by variety
plt.figure(figsize=(8, 6))
sns.scatterplot(x="sepal.length", y="petal.length", hue="variety", data=df, palette="deep", alpha=0.7)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```
